```python
def print_all_correlation(df):
    for keyc in df:
        print("\n-------- Korrelation mit %s -------" % keyc)
        corrs = [(key, np.corrcoef(df[key], df[keyc])[0][1]) for key in df if key != keyc]
        corrs.sort(key=lambda t: -t[1])
        for t in corrs:
            print("%25s %10s" % t)
print_all_correlation(bm2)
```

```python
def get_percent_table(bm):
    df = bm.copy()
    num_all = bm["Wahlberechtigte"]
    num_vote = bm["Wähler"]
    num_valid = bm["gültige Stimmen"]
    for key in df:
        if key != "Wahlberechtigte" and key != "Wähler" and key != "gültige Stimmen":
            df[key] = round(df[key] / num_valid * 100., 1)
    df["Wähler"] = round(num_vote / num_all * 100., 1)
    df["gültige Stimmen"] = round(num_valid / num_all * 100., 1)
    return df
get_percent_table(bm2)
```

```python
def get_enriched_table(bm1, bm2):
    df = bm1.copy()
    df.columns = [n.split()[-1][:4] if not "Stimmen" in n else n[:4]
                  for n in df.columns]
    
    al1 = bm1["Dr. Albrecht Schröter"]
    al2 = bm2["Dr. Albrecht Schröter"]
    th1 = bm1["Dr. Thomas Nitzsche"]
    th2 = bm2["Dr. Thomas Nitzsche"]
    
    df["a2"] = al2
    df["t2"] = th2
    df["a2-a1"] = al2 - al1
    df["t2-t1"] = th2 - th1
    df["t1-a1"] = th1 - al1
    df["t2-a2"] = th2 - al2
    df["(t2-t1)-(a2-a1)"] = (th2-th1) - (al2-al1)
    
    return df
#get_enriched_table(bm1, bm2)
```

```python
def bar_plot(series, color="#4af", back_color="#ccc"):
    mi, ma = np.min(series), np.max(series)
    fa = ma-mi
    if fa:
        fa = 1. / fa
    html = '<div style="display: flex; flex-direction: row; position:relative">'
    for i, value in enumerate(series):
        if hasattr(series, "index"):
            name = series.index[i]
        else:
            name = "%s" % i
        html += '<div style="display: block; position: relative; width: 100%%; height:70px; background: %s;" title="%s: %s">' % (color, name, value)
        html += '<div style="width:100%%; height: %s%%; background: %s;"></div>' % (round((1-(value-mi)*fa)*100, 1), back_color)
        html += '</div>'
    html += '</div>'
    from IPython.display import display, HTML
    display(HTML(html))
```

```python
bm1p = get_percent_table(bm1)
bm2p = get_percent_table(bm2)
#bar_plot(bm1p["Wähler"])
print("Schröter")
bar_plot(bm1p["Dr. Albrecht Schröter"])
bar_plot(bm2p["Dr. Albrecht Schröter"])
print("Nitzsche")
bar_plot(bm1p["Dr. Thomas Nitzsche"])
bar_plot(bm2p["Dr. Thomas Nitzsche"])
print("Nitzsche minus Schröter")
bar_plot(bm1p["Dr. Thomas Nitzsche"] - bm1p["Dr. Albrecht Schröter"])
bar_plot(bm2p["Dr. Thomas Nitzsche"] - bm2p["Dr. Albrecht Schröter"])
```

```python
def table_plot(df, color_func=None):
    if color_func is None:
        def color_func(v):
            rgb = (v*255, math.pow(v, .3)*255, math.sin(v*3.14)*255)
            return "rgb(%s, %s, %s)" % tuple((max(0, min(255, x)) for x in rgb))
                                             
    mi, ma = np.min(df.min()), np.max(df.max())
    fa = ma-mi
    if fa:
        fa = 1. / fa
    html = '<div style="display: flex; flex-direction: column; height=%spx">' % (df.shape[0] * 10)
    for y, series in enumerate(df.iterrows()):
        html += '<div style="display: flex; flex-direction: row; width=%spx">' % (df.shape[1] * 10)
        for x, value in enumerate(series[1]):            
            nvalue = (value-mi) * fa
            name = "x=%s, y=%s" % (df.columns[x], series[0])
            color = color_func(nvalue)
            html += '<div style="display: block; width: 10px; height:10px; background: %s; border: 1px solid black;" title="%s: %s">' % (color, name, value)
            html += '</div>'
        html += '</div>'
    html += '</div>'
    from IPython.display import display, HTML
    display(HTML(html))
table_plot(bm1.T)
```
