```python
def print_all_correlation(df):
    for keyc in df:
        print("\n-------- Korrelation mit %s -------" % keyc)
        corrs = [(key, np.corrcoef(df[key], df[keyc])[0][1]) for key in df if key != keyc]
        corrs.sort(key=lambda t: -t[1])
        for t in corrs:
            print("%25s %10s" % t)
print_all_correlation(bm2)
```
